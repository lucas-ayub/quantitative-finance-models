```python
import capm as capm
```

```python
stocks = ['AAPL', '^GSPC']  
start_date = '2020-01-01'
end_date = '2023-01-01'

capm = capm.Capm(stocks, start_date, end_date)

capm.initialize()

capm.calculate_beta_capm()

capm.calculate_regression()

capm.plot_parameters()

print(f"Alpha (α): {capm.alpha_reg}")
print(f"Beta (β) using CAPM: {capm.capm_beta}")
print(f"Beta (β) from regression: {capm.beta_reg}")
```
